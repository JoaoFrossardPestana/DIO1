```python
def contar_ocorrencias(lista, valor):
    """
    Função recursiva que conta ocorrências de um valor em uma lista (incluindo sublistas).
    Utiliza type() para verificar se o elemento é uma lista.
    """
    # Caso base: se a lista estiver vazia, retorna 0
    if len(lista) == 0:
        return 0

    # Separa o primeiro elemento e o restante da lista
    primeiro = lista[0]
    resto = lista[1:]

    contagem = 0

    # Uso da função type() conforme solicitado para verificar se é uma lista
    if type(primeiro) == list:
        # Se for uma lista, chama a função recursivamente para dentro dela
        contagem = contar_ocorrencias(primeiro, valor)
    elif primeiro == valor:
        # Se não for lista, verifica se é igual ao valor procurado
        contagem = 1
    else:
        # Caso contrário, não conta nada para este elemento
        contagem = 0

    # Retorna a contagem atual + o resultado da recursão no resto da lista
    return contagem + contar_ocorrencias(resto, valor)

# --- Dados fornecidos no problema ---
lista = [
    True,
    False,
    [1, 5, 8, 13],  # Contém um 8 aqui dentro
    8,
    12,
    24,
    5.25,
    12.0,
    45.8,
    3.1415,
    "caralho",
    "foda",
    "fudido",
    8
]
valor = 8

# --- Chamada da função ---
resultado = contar_ocorrencias(lista, valor)

# --- Saída ---
print(f"Lista analisada: {lista}")
print(f"Valor procurado: {valor}")
print(f"Quantidade de ocorrências: {resultado}")
```
